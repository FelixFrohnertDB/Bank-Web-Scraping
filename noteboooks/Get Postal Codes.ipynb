# SP

```python
def get_postal_sp(url_txt, save_as_csv, proxy=None):
    url_obj = open(url_txt,"r")
    url_list = [0] * len(open(url_txt).readlines())
    
    for i in range(len(url_list)):
        url_list[i] = url_obj.readline().replace('\n', '')
    
    url_list = url_list[:100]
    with open(save_as_csv, 'w', encoding='UTF8', newline='') as f:
        writer = csv.writer(f)
        
        with concurrent.futures.ThreadPoolExecutor() as executor:
            future_to_url = {executor.submit(ut.get_data_sp, url, proxy): url for url in url_list}
            for future in concurrent.futures.as_completed(future_to_url):
                try:
                    writer.writerow(future.result())
                except:
                    traceback.print_exc()
                    continue

                
    return 0
```

```python
get_postal_sp("../data/url_sp_filiale.txt", "../data/data_sp.csv")
```
